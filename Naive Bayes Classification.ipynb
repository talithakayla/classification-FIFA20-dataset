```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics 
from sklearn.metrics import classification_report
```

```
#mengimport dataset
fifa = pd.read_csv("gdrive/My Drive/classification/kmeansclustering.csv")
#manampilkan semua kolom
pd.set_option("display.max.columns", None)
#maksimal 2 angka dibelakang koma
pd.set_option("display.precision", 2)
#menampilkan sebagian data
fifa.head()
```

```
#mengecek missing values pada dataset
fifa.isna()
```

```
#mendrop atribut antara potential hingga cluster secara permanen
fifa.drop(fifa.iloc[:, 6:98], inplace=True, axis = 1) 
fifa
```

```
#mendrop semua atribut kecuali age, potential, cluster
fifa.drop(fifa.iloc[:, 0:4], inplace=True, axis = 1) 
fifa
```

```
#memasukkan nilai kelas cluster kedalam variable
fifa_df = fifa['cluster']
fifa_df
```

```
#memasukkan atribut kecuali kelas kedalam variable
fifa_la = fifa.drop(['cluster'], axis=1) 
fifa_la
```

```
#mensplit data jadi data train dan data set
x_train, x_test, y_train, y_test = train_test_split(fifa_la, fifa_df, test_size = 0.2, random_state = 0)
```

```
#melihat isi data test
x_test
```

```
#melihat isi data train
x_train
```

```
#memasukkan fungsi naive bayes dan memprediksi x_test
NB = GaussianNB() 
nbtrain = NB.fit(x_train, y_train)
y_pred = nbtrain.predict(x_test)
y_pred
```

```
#melihat hasil asli y_test
np.array(y_test)
```

```
#membuat probabilitas dari hasil x_test
nbtrain.predict_proba(x_test)
```

```
#mencari ketepatan prediksi dengan confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```
#mengeluarkan hasil akurasi
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```
#menampilkan hasil klasifikasi
print(classification_report(y_test, y_pred))
```
